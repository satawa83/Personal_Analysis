```python
import sqlite3
import os

# First, check if file exists and its size
print("Checking database file...")
print(f"File exists: {os.path.exists('data.sqlite')}")
if os.path.exists('data.sqlite'):
    print(f"File size: {os.path.getsize('data.sqlite')} bytes")

# Connect and check ALL objects
conn = sqlite3.connect('data.sqlite')
cursor = conn.cursor()

print("\nAll objects in sqlite_master:")
cursor.execute("SELECT type, name, sql FROM sqlite_master ORDER BY type;")
all_objects = cursor.fetchall()

if not all_objects:
    print("  Database is completely empty!")
else:
    for obj_type, obj_name, obj_sql in all_objects:
        print(f"  {obj_type}: {obj_name}")
        if obj_sql:
            print(f"    SQL: {obj_sql[:100]}...")  # First 100 chars

conn.close()
```
